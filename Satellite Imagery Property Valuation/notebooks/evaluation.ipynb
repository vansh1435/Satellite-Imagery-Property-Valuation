```python

# Imports & metric helpers
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
def evaluate(y_true, y_pred):
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    r2 = r2_score(y_true, y_pred)
    return rmse, r2



# Tabular-only baseline 
tabular_rmse = 63380.2764
tabular_r2 = 0.9710



# Multimodal fusion (best attempt )
fusion_rmse = 803465.81566
fusion_r2 = -0.628686


results_df = pd.DataFrame({
    "Model": ["Tabular Only", "Multimodal Fusion"],
    "RMSE": [tabular_rmse, fusion_rmse],
    "R2": [tabular_r2, fusion_r2]
})

results_df
```
